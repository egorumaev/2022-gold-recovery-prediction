### **1.2 Проверка правильности расчета эффективности обогащения**

---

Для проверки правильности расчета эффективности обогащения:

 * 1) вычислим ее на обучающей выборке для признака rougher.output.recovery

 * 2) найдем MAE (mean absolute error) между сделанными расчетами и значением признака

```python
# выведем первые пять значений столбца 'rougher.output.recovery'
# датасета train
train['rougher.output.recovery'].head()
```

Эффективность обогащения рассчитывается по формуле:

Recovery = (C * (F - T) / F*(C - T)) *100

где:

 * С - доля золота в концентрате после флотации/очистки

 * F - доля золота в сырье/концентрате до флотации/очистки

 * T - доля золота в отвальных хвостах после флотации/очистки

Соответственно для датасета train:

```python
C = train['rougher.output.concentrate_au']

F = train['rougher.input.feed_au']

T = train['rougher.output.tail_au']
```

```python
# рассчитаем эффективность обогащения. Полученные значения сохраним
# в переменную rougher.output.recovery_calculation
rougher_output_recovery_calculation = (C*(F-T))/(F*(C-T))*100
```

```python
# выведем первые пять строк переменной  rougher_output_recovery_calculation
# с расчетными значениями эффективности обогащения
rougher_output_recovery_calculation.head()
```

Рассчитаем mae между сделанными расчетами (train['rougher.output.recovery_calculation']) и значением признака (train['rougher.output.recovery'])

```python
mae = mean_absolute_error(train['rougher.output.recovery'] ,
                          rougher_output_recovery_calculation)

print('mae между расчетным значением и значением признака = ', mae)
```

#### **3.1.1. MAPE**

---

Прежде, чем перейти к вычислению итоговой метрики sMAPE, разберемся, что это за метрика. sMAPE является производной метрикой от MAPE, поэтому для ее понимания необходимо первоначлаьно подробно изучить метрику MAPE.

Средняя абсолютная процентная ошибка (MAPE, англ. Mean Absolute Percentage Error, «среднее абсолютное процентное отклонение»), также называемая средним абсолютным процентным отклонением (MAPD, англ. Mean Absolute Percentage Deviation), измеряет точность системы прогнозирования в процентах.

MAPE является наиболее распространенным показателем, используемым для оценки ошибки, поскольку выражена в процентах, что облегчает интерпретацию результатов. Использование метрики наиболее эффективно, если в данных нет крайних значений и нулей (т.е. метрика чувствительна к выбросам). Наиболее часто MAPE используется при оценке модели в регрессионном анализе.

При расчете MAPE берется модуль от разницы целевого и предсказанного значений, что позволяет устранить проблему положительных и отрицательных значений ошибок, компенсирующих друг друга.

 * Mean Absolute Percentage Error (MAPE) // Statistics How To. – URL: https://www.statisticshowto.com/mean-absolute-percentage-error-mape/

Итак, MAPE возвращает ошибку в процентах, что позволяет оценить правильность прогноза модели. Общий устоявшийся на практике подход к интерпретации MAPE следующий:

 * MAPE < 10%: отличный результат

 * MAPE от 10 до 20%: хороший результат

 * MAPE от 20 до 50%: удовлетворительный результат

 * MAPE > 50%: неудовлетворительный результат

Хорошее значение метрики MAPE находится в диапазоне менее 20%. Подобный результат означает, что в среднем прогнозы за весь период времени отличаются от фактических значений менее, чем на 20%.

Значение MAPE могут превышать 100%, поскольку предсказания могут иметь ошибку, превышающую фактическое значение. Однако, MAPE – это показатель ошибки, значение которого соответствует среднему значению ошибки, присущей прогнозам. Следовательно, целесообразно использовать модель с меньшим значением MAPE: чем меньше значение, тем точнее модель. Значение MAPE варьируется от 0 до бесконечности, где чем меньше значение, тем точнее прогнозы.

 * What is a good MAPE score? // Stephen Allwright. – URL: https://stephenallwright.com/good-mape-score/

MAPE можно интерпретировать как «обратную точность модели», – как среднюю процентную разницу между прогнозами и их предполагаемыми целями в наборе данных. Например, если MAPE составляет 10%, то прогнозы в среднем на 10% отличаются от фактических значений, к которым они стремились.

 * How to interpret MAPE (simply explained) // Stephen Allwright. – URL: https://stephenallwright.com/interpret-mape/

sMAPE часто используется вместо MAPE, чтобы избежать работы с метрикой неограниченного размера: как было отмечено выше, значение MAPE может быть от 0 до бесконечности. Однако, в ряде случаев возможно интерпретировать sMAPE точно так же, как рассчитанный с помощью MAPE процент ошибок.

В задачах машинного обучения выбор основной метрики, оценивающей качество обученной модели, является не простой задачей. Поэтому часто прибегают к измерению точности относительно актуальных (целевых) значений. И в этой части самой популярной метрикой является MAPE.

У использования метрики MAPE есть строгое ограничение: она трансформируется в положительную бесконечность, как только любое из актуальных (целевых) значений становится равным нулю:

**Преимущества** метрики MAPE:

 * выражается в %, которые не зависят от масштаба и могут быть использованы для сравнения прогнозов в разных масштабах (для разных моделей). Однако следует помнить, что значения MAPE могут превысить 100%

 * метрика доступна для понимания всем заинтересованным сторонам проекта

**Ограничения** использования метрики MAPE:

 * MAPE принимает неопределенные значения, когда фактические значения равны нулю. Это нередко происходит при прогнозировании бизнес-показателей, например, спроса. Кроме того, метрика принимает экстремально большие значения, когда фактически значения близки, но не равны нулю

 * Метрика MAPE ассимметрична. Это выражается в том, что при ее расчете налагается более высокое «наказание» за отрицательные ошибки (когда прогнозы превышают фактические данные), чем за положительные ошибки. Это объясняется тем, что процентная ошибка не может превышать 100% для прогнозов, которые являются слишком низкими. В результате применение метрики MAPE предпочтительно для моделей с заниженным, а не завышенным прогнозом

 * MAPE не применима при прогнозировании явлений, имеющих произвольную нулевую точку (например, прогнозирование температуры, вероятности нулевых выплат по страховкам)

#### **3.1.2 Использование метрики MAPE в бизнесе**


---

Компании часто используют прогноз для прогнозирования того, что они собираются продавать. Существует ряд показателей, которые компании используют для измерения точности прогноза. MAPE широко применяется в оценке качества предсказаний, полученных с помощью машинного обучения.

#### **3.1.3. sMAPE**

---


Использование MAPE ограничено случаями, если все наблюдения имеют относительно большие фактические значения. В обратном случае можно использовать аналог MAPE, которым является sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).

Cуществует два определения sMAPE.

**sMAPE. Определение первое**: нормализует относительные ошибки путем деления как на фактические, так и на прогнозируемые значения. В результате показатель находится в диапазоне от 0% до 100%:

Формула sMAPE 1.JPG

**sMAPE. Определение второе**: второе определение похоже на первое, за исключением деления на два (2) в знаменателе. В этом случае диапазон значений sMAPE расширяется от 0% до 200%:

Формула sMAPE 2.JPG

Более простым представляется путь, когда sMAPE интерпретируется аналогично MAPE. Но этот подход ошибочен.

Поскольку ошибки для «первого» sMAPE распределены в диапазоне от 0% до 100%, эти показатели интуитивно понятны. Однако результаты расчета значения «второго» sMAPE не столь очевидны.

Малые значения «второго» sMAPE (до 30%) могут быть интерпретированы точно так же, как и значения MAPE. Аналогичным образом могут быть интерпретированы значения «первого» sMAPE, если их умножить на 2 (два).

 * Sarra, D. How to interpret sMAPE just like MAPE / D. Sarra // Medium – Where good ideas find you. – URL: https://medium.com/@davide.sarra/how-to-interpret-smape-just-like-mape-bf799ba03bdc

**Преимущества** метрики sMAPE:

 * выражена в процентах

 * исправлен основной недостаток метрики MAPE, а именно: есть нижняя граница (0%) и верхняя граница (200%)

**Ограничения** использования метрики sMAPE:

 * нестабильна, когда фактическое и прогнозное значение близки к нулю: появляется риск деления на ноль

 * может принимать отрицательные значения, поэтому интерпретация может вводить в заблуждение

 * диапазон от 0% до 200% не очень интуитивно понятен, поэтому нередко деление на 2 (два) в знаменателе формулы опускают

 * при достижении фактического или прогнозного значения, равного нулю (0), метрика автоматически достигает верхнего граничного значения (100%)

 * устраняя асимметрию, характерную для MAPE, метрика sMAPE подвержена другому виду асимметрии. Например, при А = 100 и В = 120 sMAPE = 18.2%. При очень похожем случае, когда А = 100 и В = 80 метрика sMape = 22.2%

Lewinson, E. Choosing the correct error metric: MAPE vs. sMAPE / E. Lewinson // Towards Data Science. – URL: https://towardsdatascience.com/choosing-the-correct-error-metric-mape-vs-smape-5328dec53fac

Таким образом, в целом чем ниже sMAPE сделанного прогноза, тем выше его точность. Поскольку метрика ориентирована на расчет процента ошибки, она не зависит от масштаба, что открывает возможность проводить сравнение прогнозов, полученных на разных наборах данных.

Однако в ряде случаев оценивать результат расчета метрики необходимо с осторожностью: при наличии фактических или прогнозируемых значений, равных нулю, значение ошибки приближается к 100% и в этом случае не отображает реальное качество прогноза.

 * Understand advanced metrics // Anaplan. – URL: https://help.anaplan.com/685ff9b2-6370-46ba-af10-679405937113-Understand-advanced-metrics

### **3.2 Подготовка функции для вычисления sMAPE и итоговой метрики sMAPE**

---

Согласно технического задания необходимо рассчитать:

 * метрику sMAPE для эффективности обогащения чернового концентрата (1) на этапе флотации (целевой признак 'rougher.output.recovery') и (2) на финальном этапе (целевой признак 'final.output.recovery') по формуле:

Формула.JPG

 * итоговую метрику sMAPE по формуле:

Формула.JPG

Создадим функции для расчета sMAPE и sMAPE_final. sMAPE будет рассчитано в двух вариантах: по вышеуказанной формуле и с помощью функции из библиотеки torchmetrics. Функция будет возвращать оба значения. Для итогового sMAPE будет взят результат, рассчитанный по формуле. Второй вариант, рассчитываемый с помощью функции из библиотеки torchmetrics, определяется в порядке эксперимента.

```python
# создадим функцию для расчета метрики sMAPE для оценки эффективности обогащения
# чернового концентрата на этапе флотации и на финальном этапе
def smape_calculation(target, predicted):

    # считаем sMAPE по формуле
    smape_by_formula = round(
        np.mean(
            np.abs(predicted - target) /
            (    (np.abs(predicted) + np.abs(target)) / 2)
                )*100, 2
    )

    # считаем sMAPE с помощью функции из библиотеки torchmetrics

    # target и predicted переводим в формат tensor
    target = torch.tensor([target])
    predicted = torch.tensor([predicted])

    # создаем объект SymmetricMeanAbsolutePercentageError
    smape = SymmetricMeanAbsolutePercentageError()

    # вычисляем sMAPE с помощью функции из библиотеки torchmetrics
    smape_by_function = smape(predicted, target)

    return smape_by_formula, smape_by_function * 100
```

```python
# создадим функцию для расчета итоговой метрики sMAPE
def smape_final_result(smape_rougher_stage, smape_final_stage):
    smape = 0.25 * smape_rougher_stage + 0.75 * smape_final_stage
    return smape
```

 * расcчитанного по стандартной формуле

 * рассчитанного с помощью функции из библиотеки torchmetrics

#### **3.3.1 Создание собственной метрики с помощью make_scorer**

---

Для того, чтобы при обучении моделей использовать кросс-валидацию, необходима переменная для указания в параметре scoring. В библиотеке sklearn нет метрики для sMAPE, поэтому подготовим ее самостоятельно с помощью инструмента make_scorer библиотеки sklearn. Для этого потребуется первая часть функции smape_calculation из п. 3.2. Мы могли бы использовать ее целиком, но в связи с проведением эксперимента по расчету sMAPE функцией из библиотеки torchmetrics применение функции в полном варианте вызовет ошибку при проведении кроссвалидации.

```python
# используем часть функции smape_calculation
# для расчета метрики sMAPE
# для создания собственной метрики при помощи make_scorer
def our_own_smape(target, predicted):

    # считаем sMAPE по формуле
    smape = round(
        np.mean(
            np.abs(predicted - target) /
            (    (np.abs(predicted) + np.abs(target)) / 2)
                )*100, 2
    )

    return smape
```
